```python
zeros = np.zeros(5)
ones = np.ones((3, 3))
print("zeros:", zeros)
print("ones:\n", ones)
```

```python
# Changing shape
big_zeros = np.zeros(10)
reshaped = big_zeros.reshape(10, 1)

print("reshaped:\n", reshaped)
```

```python
# Empty array
empty_arr = np.empty((2, 2))
print("empty_arr:\n", empty_arr)
```

```python
# Linspace
linear = np.linspace(2, 10, 5)
print("linear:\n", linear)
```

zeros → creates array filled with 0
ones → creates matrix of 1s
reshape → changes shape (important for ML)
empty → fast but contains garbage values
linspace → evenly spaced numbers
array() → converts list → NumPy array
Why:
Used for initialization, simulations, ML inputs
